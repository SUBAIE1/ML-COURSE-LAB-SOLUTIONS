# Lab 3: Exploratory Data Analysis (EDA)

EDA is the first and most important step in any Machine Learning project.
Before building models, we must understand:

- What does the data represent?
- Are there missing values?
- Are there outliers?
- What patterns exist?
- Which variables influence others?

If we do not understand the data, we cannot build a good model.

### Why EDA is Important

In real-world machine learning projects:

    1- 70–80% of the time is spent on understanding and cleaning data
    2- Only 20–30% is spent building models

**EDA helps us:**

- Detect errors
- Identify trends
- Discover relationships
- Make business decisions

## Bivariate Analysis.

### Quantity_Sold by Product_Category

```python
Quantity_Sold_Product_Category = df.groupby('Product_Category')['Quantity_Sold'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
Quantity_Sold_Product_Category.plot(kind='bar')
plt.title("Quantity soled per product category")
plt.ylabel("Quantity")
plt.show()

Quantity_Sold_Product_Category
```

```python
sales_per_Product_Category = df.groupby('Product_Category')['Sales_Amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
Quantity_Sold_Product_Category.plot(kind='bar')
plt.title("Sales per product category")
plt.ylabel("Sales Amount")
plt.show()

Quantity_Sold_Product_Category
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Discount', y='Sales_Amount', data=df)
plt.title("Discount Vs Sales_Amount")
plt.show()


```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['Discount', 'Sales_Amount']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
df['Month'] = df['Sale_Date'].dt.to_period('M')

monthly_revenue = df.groupby('Sale_Date')['Sales_Amount'].sum()

plt.figure(figsize=(10,5))
monthly_revenue.plot()
plt.title("Monthly Sales Amount Trend")
plt.ylabel("Salse Amount")
plt.show()
```
